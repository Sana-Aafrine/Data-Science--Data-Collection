```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
country_listing=pd.DataFrame()
r=1
First=True
for j in range(1,51):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    if First==True:
        url="https://www.realtor.com/international/in"
        webpage=requests.get(url,headers=headers).text
        soup=BeautifulSoup(webpage,'lxml')
        First=False
    else:
        url="https://www.realtor.com/international/in/p{}".format(j)
        webpage=requests.get(url,headers=headers).text
        soup=BeautifulSoup(webpage,'lxml')
    listing = soup.find_all('div',class_='sc-1dun5hk-5')
    listing_price_dollars = []
    listing_price_rupees = []
    address = []
    bedrooms = []
    washrooms = []
    sq_feet_area =[]
    type_of_listing = []
    for i in listing:
        try:
            listing_price_dollars.append(i.find('div',class_='displayConsumerPrice').text.strip())
        except:
            listing_price_dollars.append(np.nan)
        try:
            address.append(i.find('div',class_='address').text.strip())
        except:
            address.append(np.nan)
        try:
            bedrooms.append(i.find_all('div',class_='feature-item')[0].text.strip())
        except:
            bedrooms.append(np.nan)
        try:
            washrooms.append(i.find_all('div',class_='feature-item')[1].text.strip())
        except:
            washrooms.append(np.nan)
        try:
            sq_feet_area.append(i.find_all('div',class_='feature-item')[2].text.strip())
        except:
            sq_feet_area.append(np.nan)
        try:
            type_of_listing.append(i.find('div',class_='property-type').text.strip())
        except:
            type_of_listing.append('Unit')
    page=pd.DataFrame({'Listing Price' : listing_price_dollars,
                          'Address' : address,
                          'Number of Bedrooms' : bedrooms,
                          'Number of Bathrooms' : washrooms,
                          'Area(sq-ft.)' : sq_feet_area,
                          'Property Type' : type_of_listing})
    country_listing=pd.concat([country_listing,page],ignore_index=True)
```

```python
country_listing.sample(15)
```

```python
country_listing.to_csv("India_property_listing.csv")
```
