```python
df=response_df
df.info()
```

```python
df['tests'][27]
```

```python
total=[]
death=[]
tests=[]
recovered=[]
for i,j in df.iterrows():
    total.append(j['cases']['total'])
    death.append(j['deaths']['total'])
    recovered.append(j['cases']['recovered'])
    tests.append(j['tests']['total'])
    
df['Cases'] = total
df['Recovered'] = recovered
df['Deaths'] = death
df['Tests'] = tests
```

```python
df=df.set_index('country')
```

```python
df=df.drop(columns=['cases','tests','deaths'])
```

```python
df.to_csv('covid_19.csv')
```
