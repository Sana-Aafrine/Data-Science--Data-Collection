## 3.Opening a CSV File from an URL

```python
import requests
from io import StringIO

url="https://drive.google.com/uc?id=1zO8ekHWx9U7mrbx_0Hoxxu6od7uxJqWw&export=download"
headers= {"User-Agent": "Mozilla/5.0 (Machintosh; Intel MAC OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req= requests.get(url, headers=headers)
data= StringIO(req.text)

pd.read_csv(data)
```

## 5. 'index_col' Parameter
We can make a column the index column throught this parameter

```python
pd.read_csv('customers-100.csv',index_col='Index')
```

## 6.'header' Parameter
If the column names is not being recognised and added to data values, we can specify header parameter as (header=1) to take the first row of values as the column names.

## 7.'usecols' Parameter
We can specify the required columns to be included in the dataset by specifying the columns names that are to be extracted.

```python
pd.read_csv('customers-100.csv',index_col='Index',usecols=['Index','Customer Id','First Name','Phone 1'])
```

## 8.'skiprows'/'nrows'
We can specify which rows to skip y skiprows and limit no of rows by nrows

```python
pd.read_csv('customers-100.csv',index_col='Index', nrows=10, skiprows=[1,3,5])
```

## 9.'encoding' Parameter
Some datasets are not utf-8 encoded and might throw errors. They either need to be encoded to utf-8 or their encoding values need to be found out and specified.

## 10.Skip Bad Lines
Often if there are abruptly more values in a row than the number of columns, an error occurs. It can be fixed by making the 'error_bad_lines' parameter to False.

## 12.Handling Dates
Dates are often handled as objects in python. We need to speccify the 'parse_date' parameter to read the columns as dates.

```python
pd.read_csv('customers-100.csv',index_col='Index').info()
```

```python
pd.read_csv('customers-100.csv',index_col='Index',parse_dates=['Subscription Date']).info()
```

## 13.Convertors
To convert values present in the dataset column to different set of values according to a function.

```python
def rename(name):
    if name=="BIRDS":
        return "B"
    elif name=="THE LIBRARY":
        return "TL"
    else:
        return name

pd.read_csv('kids_teen.tsv', sep='\t', converters={'category':rename})
```

## 14.'na_values' Parameter
Data inputs like '-' also indicate missing values but are often not identified and therefore need to be exclusively specifies using 'na_values' parameters.

```python
pd.read_csv('kids_teen.tsv', sep='\t',na_values=['-']).head()
```

## 15.Loading huge dataset in chunks
If the dataset is very large and cannot be processed as a whole, we load the dataset in chunks and perform operations using loops.

```python
ch=pd.read_csv('kids_teen.tsv', sep='\t',chunksize=6000)
```

```python
for chunks in ch:
    print(chunks.size)
```
